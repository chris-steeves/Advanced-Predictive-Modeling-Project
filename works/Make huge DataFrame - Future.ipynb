```python
import pandas as pd
from pandas import DataFrame, Series
```

```python
all_games = pd.read_csv('allgames_future_v6.csv')
```

```python
all_games
```

```python
#added to get rid of future week   all_games = all_games[all_games['W/L'] != '-']
all_games = all_games.replace('-',0).fillna(0)
```

```python
#sort values

all_games['Week'] = all_games['Week'].replace(['WC', 'DP', 'CC', 'SB'], [18, 19, 20, 21])
all_games['Week'] = all_games['Week'].astype(int)
all_games.sort_values(['Team', 'Season', 'Week'], ascending = [True, True, True], inplace = True)
all_games.reset_index(inplace = True)
```

```python
all_games
```

```python
attributes = ['OVR', 'OFF', 'PASS', 'PBLK', 'RECV', 'RUN', 'RBLK', 'DEF', 'RDEF', 'TACK', 'PRSH', 'COV']
```

```python
import copy

lagged = copy.deepcopy(all_games)
```

```python
lagged['Team16'] = lagged['Team'].shift(16)

for attribute in attributes:
    
    for i in range(1,17):
        col = attribute + '_lag' + str(i)
        lagged[col] = lagged[attribute].shift(i)
    
    avg_col = attribute + '_avg3'
    
    lagged[avg_col] = (lagged[attribute + '_lag1'].astype(float) + 
                       lagged[attribute + '_lag2'].astype(float) + 
                       lagged[attribute + '_lag3'].astype(float)) / 3

lagged = lagged[lagged['Team'] == lagged['Team16']]
    
```

```python
final = lagged.merge(lagged,
                     left_on = ['GameNum', 'Team'],
                     right_on = ['GameNum', 'Opposing Team'],
                     suffixes = ['', '_opposing'])
```

```python
removable_cols = ['index', 'Unnamed: 0', 'Opposing Team',  'H/A', 'Date_opposing', 'Time_opposing', 'W/L_opposing', 'PF_opposing', 'PA_opposing', 'Week_opposing', 'Season_opposing', 'index_opposing', 'Unnamed: 0_opposing', 'Opposing Team_opposing',  'H/A_opposing', 'Team16', 'Team16_opposing']
```

```python
final.drop(columns = removable_cols, inplace = True)
```
